```python
# LIbraries needed
from faker import Faker
import pandas as pd


fake = Faker()
def generate_fake_data(num_entries=100):
    data={
        'Name':[],
        'Credit card number':[],
        'Credit card provider':[],
        'Currency':[],
        'Phone Number':[],
        'Email':[],
    }

    for _ in range(num_entries):
        data['Name'].append(fake.name())
        data['Credit card number'].append(fake.credit_card_number())
        data['Credit card provider'].append(fake.credit_card_provider())
        data['Currency'].append(fake.currency())
        data['Phone Number'].append(fake.phone_number())
        data['Email'].append(fake.email())
        
    return data
    
#Generate the data
fake_data = generate_fake_data(100)

#Create a DataFrame
df_fake_data = pd.DataFrame(fake_data)

#Save to a CSV file
df_fake_data.to_csv(r"fake_data_finance1.csv", index=False)
df_fake_data.head()
```
